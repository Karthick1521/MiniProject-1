#2.Data Extraction & Transformation (SQL & Python):

```python
#DB connection:
import mysql.connector
import pandas as pd

# Connect to MySQL database
db_connection = mysql.connector.connect(
    host="localhost",      
    user="root",            
    password="@123Qweasz",    
    database="customer_behaviour_analysis" 
)

curr = db_connection.cursor()
print("Database connected successfully!")
```

```python
#1.Extract Relevant Data:
#Get customer details and their product reviews.

query = """
SELECT c.CustomerID, c.CustomerName, cr.ProductID, cr.Rating, cr.ReviewDate
FROM customers c
JOIN customer_reviews cr ON c.CustomerID = cr.CustomerID;
"""
curr.execute(query)
result = curr.fetchall()

# Convert to DataFrame
df = pd.DataFrame(result, columns=['CustomerID', 'CustomerName', 'ProductID', 'Rating', 'ReviewDate'])
print(df.head())
```

```python
#2.Perform Joins to Integrate Datasets
#Fetch customer journeys along with product details and engagements.

query = """
SELECT c.CustomerID, c.CustomerName, p.ProductName, ed.ViewsClicksCombined, ed.EngagementDate
FROM customers c
JOIN customer_journey cj ON c.CustomerID = cj.CustomerID
JOIN products p ON cj.ProductID = p.ProductID
JOIN engagement_data ed ON p.ProductID = ed.ProductID;
"""
curr.execute(query)
result = curr.fetchall()

df = pd.DataFrame(result, columns=['CustomerID', 'CustomerName', 'ProductName', 'ViewsClicksCombined', 'EngagementDate'])
print(df.head())
```

```python
#3.Apply window functions, CTEs, and subqueries for deeper insights.
LEAD Query to get the next visit date
lead_query = """
    SELECT CustomerID, VisitDate,
           LEAD(VisitDate) OVER (PARTITION BY CustomerID ORDER BY VisitDate) AS NextVisitDate
    FROM customer_journey;
"""

curr.execute(lead_query)
lead_data = curr.fetchall()

# Convert to DataFrame
lead_df = pd.DataFrame(lead_data, columns=['CustomerID', 'VisitDate', 'NextVisitDate'])
print("LEAD Query Results:")
print(lead_df.head(10))  
```

```python
# LAG Query to get the previous visit date
lag_query = """
    SELECT CustomerID, VisitDate,
           LAG(VisitDate) OVER (PARTITION BY CustomerID ORDER BY VisitDate) AS PreviousVisitDate
    FROM customer_journey;
"""

curr.execute(lag_query)
lag_data = curr.fetchall()

# Convert to DataFrame
lag_df = pd.DataFrame(lag_data, columns=['CustomerID', 'VisitDate', 'PreviousVisitDate'])
print("LAG Query Results:")
print(lag_df.head(10)) 
```

```python
# RANK Query to rank products based on Rating
rank_query = """
    SELECT ProductID, CustomerID, Rating,
           RANK() OVER (PARTITION BY ProductID ORDER BY Rating DESC) AS RankPosition
    FROM customer_reviews;
"""

curr.execute(rank_query)
rank_data = curr.fetchall()

# Convert to DataFrame
rank_df = pd.DataFrame(rank_data, columns=['ProductID', 'CustomerID', 'Rating', 'RankPosition'])
print("RANK Query Results:")
print(rank_df.head(10))  
```

```python
# DENSE RANK Query to rank products with no rank gaps
dense_rank_query = """
    SELECT ProductID, CustomerID, Rating,
           DENSE_RANK() OVER (PARTITION BY ProductID ORDER BY Rating DESC) AS DenseRankPosition
    FROM customer_reviews;
"""

curr.execute(dense_rank_query)
dense_rank_data = curr.fetchall()

# Convert to DataFrame
dense_rank_df = pd.DataFrame(dense_rank_data, columns=['ProductID', 'CustomerID', 'Rating', 'DenseRankPosition'])
print("DENSE RANK Query Results:")
print(dense_rank_df.head(10)) 
```

```python
#This Query finds the most recent purchase of each customer using a CTE.

cte_query = """
    WITH RecentPurchase AS (
        SELECT CustomerID, MAX(VisitDate) AS LastPurchaseDate
        FROM customer_journey
        WHERE Action = 'Purchase'
        GROUP BY CustomerID
    )
    SELECT r.CustomerID, cu.CustomerName, r.LastPurchaseDate
    FROM RecentPurchase r
    JOIN customers cu ON r.CustomerID = cu.CustomerID;
"""

curr.execute(cte_query)
data = curr.fetchall()

df = pd.DataFrame(data, columns=['CustomerID', 'CustomerName', 'LastPurchaseDate'])
print("Recent Purchases:")
print(df)
```

```python
#4.Customer Reviews Analysis (Python):
#2.Perform basic sentiment analysis in Python
```

```python
import pandas as pd

# Word lists for sentiment analysis
positive_words = ['good', 'great', 'excellent', 'amazing', 'love', 'happy', 'satisfied', 'awesome']
negative_words = ['bad', 'poor', 'terrible', 'hate', 'awful', 'unhappy', 'worst', 'disappointed']

# Expanding word lists to improve sentiment analysis
positive_words.extend(['five stars', 'love it', 'amazing service', 'exceeded expectations'])
negative_words.extend(['not worth', 'disappointed', 'horrible experience', 'waste of money'])

# Function to analyze sentiment
def analyze_sentiment(text):
    if pd.isna(text):
        return 'Neutral'
    
    text = text.lower()

    # Prioritize matching complete phrases if available
    for phrase in positive_words:
        if phrase in text:
            return 'Positive'
    for phrase in negative_words:
        if phrase in text:
            return 'Negative'
    
    return 'Neutral'

# Load data from CSV
customer_reviews = pd.read_csv("customer_reviews.csv")

# Ensure ReviewText column exists
if 'ReviewText' not in customer_reviews.columns:
    raise KeyError("Column 'ReviewText' not found in DataFrame")

# Apply sentiment analysis
customer_reviews['Sentiment'] = customer_reviews['ReviewText'].apply(analyze_sentiment)

# Display results
print(customer_reviews[['ReviewText', 'Sentiment']].head(100))



```

```python
#3.Correlate review trends with product performance.
```

```python
import pandas as pd

# Load the customer reviews and product data
customer_reviews = pd.read_csv("customer_reviews.csv")
products = pd.read_csv("products.csv")

avg_ratings = customer_reviews.groupby('ProductID')['Rating'].mean().reset_index()
print("Average Ratings per Product:")
print(avg_ratings)


positive_words = ['good', 'great', 'excellent', 'amazing', 'love', 'happy', 'satisfied', 'awesome']
negative_words = ['bad', 'poor', 'terrible', 'hate', 'awful', 'unhappy', 'worst', 'disappointed']

def analyze_sentiment(text):
    if pd.isna(text):
        return 'Neutral'
    
    text = text.lower()
    
    # Simple sentiment analysis
    if any(word in text for word in positive_words):
        return 'Positive'
    elif any(word in text for word in negative_words):
        return 'Negative'
    else:
        return 'Neutral'
        
customer_reviews['Sentiment'] = customer_reviews['ReviewText'].apply(analyze_sentiment)

# Count how many Positive, Negative, and Neutral reviews each product has
sentiment_count = customer_reviews.groupby(['ProductID', 'Sentiment']).size().reset_index(name='Count')
print("\nSentiment Count per Product:")
print(sentiment_count)

# Merge product details with average rating and sentiment count
product_performance = pd.merge(products, avg_ratings, on='ProductID', how='left')
product_performance = pd.merge(product_performance, sentiment_count, on='ProductID', how='left')

print("\nFinal Product Performance Data:")
print(product_performance)

```
